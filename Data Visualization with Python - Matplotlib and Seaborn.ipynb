# 7. PLOT 3D PLOTS

```python
# Toolkits are collections of application-specific functions that extend Matplotlib.
# mpl_toolkits.mplot3d provides tools for basic 3D plotting.
# https://matplotlib.org/mpl_toolkits/index.html

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (20, 15))
ax = fig.add_subplot(111, projection = '3d')

x = [10, 4, 3, 8, 5, 13, 1, 6, 9, 10]
y = [8, 6, 21, 0, 13, 4, 9, 2, 3, 1]
z = [4, 10, 7, 3, 5, 14, 9, 1, 12, 1]

ax.scatter(x,y,z,c='r',s=500)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

```

# 8. SEABRON SCATTERPLOT & COUNTPLOT

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()

# Import Cancer data drom the Sklearn library
```

```python
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the tail of the dataframe
df_cancer.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.figure(figsize = (10,10))
sns.scatterplot(x='mean radius',y='mean texture',hue='target',data=df_cancer)
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
plt.figure(figsize = (10,10))
sns.countplot(df_cancer['target'],label='count')
```

MINI CHALLENGE #8:
- Plot the scatterplot between the mean radius and mean area. Comment on the plot


```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='mean radius',y='mean area',hue='target',data=df_cancer)
```

# 9. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS


```python
df_cancer.head()
```

```python
# Plot the pairplot
sns.pairplot(df_cancer,hue='target',vars=['mean radius','mean area','mean perimeter','mean symmetry','mean compactness','mean smoothness'])
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize = (30, 30)) 
sns.heatmap(df_cancer.corr(),annot=True)
```

```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 
plt.figure(figsize=(20,10))
sns.distplot(df_cancer['mean radius'],bins=20,color='y')
```

```python
plt.figure(figsize=(20,10))
sns.distplot(df_cancer_class0['mean area'],bins=30,color='b')
sns.distplot(df_cancer_class1['mean area'],bins=30,color='r')
plt.grid()
```
